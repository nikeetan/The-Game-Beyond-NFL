```python
import pandas as pd


df = pd.read_csv('data/imputed_data.csv')

df['Position'] = df['Position'].astype('category').cat.codes


df.drop(columns=['Name','Stat URL','Year','College','conf_abbr'], axis=1, inplace=True)
df
```

```python
import seaborn as sns
from IPython.core.display import display, HTML

import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr = df.corr()

# Generate a heatmap
plt.figure(figsize=(22, 22))
ax = sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
ax.xaxis.set_ticks_position('top')
plt.xticks(rotation=90)


plt.show()
```
